```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# Importing dataset
data = pd.read_csv('Diwali_Sales_Data.csv', encoding='latin1')
```

```python
# Taking out random data
data.sample(5)
```

```python
# Information about each columns
data.info()
```

```python
# Information about roas and columns in dataset
data.shape
```

```python
# Dropping the columns which doesn't hold any data
data.drop(['Status','unnamed1'],inplace = True,axis = 1)
```

```python
data.info()
```

```python
# Column wise null values
pd.isnull(data).sum()
```

```python
# Removing the null data
data.dropna(inplace = True)
```

```python
data.info()
```

```python
# Changing datatypes
data['Amount'] = data['Amount'].astype('int')
```

```python
# Checking data types after changing
data['Amount'].dtypes
```

```python
data.columns
```

```python
# Renaming the column name from Marital_Status to Shaadi
data.rename(columns = {'Marital_Status':'Shaadi'},inplace = True)
```

```python
data
```

```python
data.describe()
```

```python
# Displaying the Amount,Orders And Age information
data[['Amount','Orders','Age']].describe()
```

### Exploratory Data Analysis

```python
# Gender Distribution
plt.style.use('dark_background')

ax = sns.countplot(x = 'Gender',data = data,hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# grouping Gender Distribution In Terms Of Purchasing
sales_gen = data.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen,hue = 'Gender')
```

### Age

```python
data.columns
```

```python
# Age Group Of Purchaser
ax = sns.countplot(data = data,x = 'Age Group',hue = 'Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Group by on Age group in terms of amount
sales_age = data.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
# Bar plot of age group
sns.barplot(x = 'Age Group',y = 'Amount',data = sales_age,hue = 'Age Group')
```

### State

```python
data.columns
```

```python
# Total numbers of orders from top 10 states
sns.set_style("dark")  # or "dark", "whitegrid", "white", "ticks"
sales_state = data.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x = 'State',y = 'Orders')
```

```python
# Total numbers of amount/sales from top 10 states
sales_state = data.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x = 'State',y = 'Amount')
```

### Marital Status

```python
# Distribution of married and un-married
ax = sns.countplot(data = data,x = 'Shaadi')
sns.set(rc = {'figure.figsize':(7,5)})

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Gender distribution among married and un-married people
sales_state = data.groupby(['Shaadi', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Shaadi',y= 'Amount', hue='Gender')
```

### Occupation

```python
# Occupation of customers and their distribution
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = data,x = 'Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Top purchasers
sales_state = data.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x = 'Occupation',y = 'Amount')
```

### Product Category

```python
# Quantity of products category
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = data,x = 'Product_Category')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Products sales
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',hue = 'Product_Category')
```

### Products sold

```python
# Products sold with their ID
sales_state = data.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x = 'Product_ID',y = 'Orders')
```

```python
# top 10 most sold products (same thing as above)
fig1, ax1 = plt.subplots(figsize=(12,7))
data.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
